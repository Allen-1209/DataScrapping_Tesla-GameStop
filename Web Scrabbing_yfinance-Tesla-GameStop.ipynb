```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
```

```python
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data, 'html5lib')
```

```python
tesla_revenue = pd.DataFrame(columns=["Year","Revenue"])
tesla_revenue
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Year = col[0].text
    Revenue = col[1].text
    
    tesla_revenue = tesla_revenue.append({"Year":Year,"Revenue":Revenue}, ignore_index=True)    
```

```python
tesla_revenue.tail()
```

```python
gme = yf.Ticker("GME")
gme_data=gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
url_gme="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
```

```python
data_gme  = requests.get(url_gme).text
soup = BeautifulSoup(data_gme, 'html5lib')
gme_revenue = pd.DataFrame(columns=["Year","Revenue"])
gme_revenue
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Year = col[0].text
    Revenue = col[1].text
    gme_revenue = gme_revenue.append({"Year":Year,"Revenue":Revenue}, ignore_index=True) 
```

```python
gme_revenue.tail()
```

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    stock_data_specific = stock_data[stock_data.Date <= '2021--06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Year, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Year, infer_datetime_format=True), y=revenue_data_specific.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Year", row=1, col=1)
    fig.update_xaxes(title_text="Year", row=2, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=2)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
make_graph(gme_data, gme_revenue, 'GameStop')
```
